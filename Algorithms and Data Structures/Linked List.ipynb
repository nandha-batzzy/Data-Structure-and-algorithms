# Linked List

```python
class Node:
    def __init__(self,data = None,next = None):
        self.data = data
        self.next = next
class llist:
    def __init__(self):
        self.head = None
    
    def insert(self, data):
        node = Node(data, self.head)
        self.head = node
    def print(self):
        if self.head is None:
            print("linked list is empty")
            return
        
        itr = self.head
        lstitr = ""
        while itr:
            lstitr = lstitr + str(itr.data) + "-->"
            itr = itr.next
        print(lstitr)
    
    def insert_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data,None)
            

if  __name__ == '__main__':

    lis = llist()
    lis.insert(3)
    lis.insert(66)
    lis.insert("nandha")
    lis.insert("genius")
    lis.insert(89)
    lis.insert_end(13)
    lis.insert_end(77)
    lis.print()
```

```python
class Node:
    def __init__(self,data = None,next = None):
        self.data = data
        self.next = next
class llist:
    def __init__(self):
        self.head = None
    
    def insert_beg(self, data):
        node = Node(data, self.head)
        self.head = node
        
    def insert_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data,None)
        
    def print(self):
        if self.head is None:
            print("linked list is empty")
            return
        
        itr = self.head
        lstitr = ""
        while itr:
            lstitr = lstitr + str(itr.data)  
            itr = itr.next
        return lstitr
        
    def printrev(self):
        if self.head is None:
            print("linked list is empty")
            return
        
        itr = self.head
        lstitr = ""
        while itr:
            lstitr =str(itr.data) + lstitr 
            itr = itr.next
        return lstitr    
        
    def insert_values(self,dataset):
        self.head = None
        for i in dataset:
            self.insert_end(i)
    
    def count(self):
        c = 0
        itr = self.head
        while itr:
            c = c+1
            itr = itr.next
        return c 
    def remove(self,index):
        if index<0 or index>=self.count():
            raise Exception("invalid index")
        if index == 0:
            self.head = self.head.next
            return
        c = 0 
        itr = self.head
        while itr:
            if c == index -1:
                itr.next = itr.next.next
            itr = itr.next
            c = c+1
    def insert_aft_value(self,value,data):
        if self.head is None:
            self.insert_beg(data)
            return 
        itr = self.head
        while itr:
            if itr.data == value:
                itr.next = Node(data,itr.next)
                break
            itr = itr.next
        return
    def remove_by_value(self,data1):
        itr = self.head
        while itr.next:
            if itr.next.data == data1:
                itr.next = itr.next.next
                break
            itr = itr.next
        return
    def reverse(self):
        previous = None
        current = self.head     
        following = current.next    


        while current:
            current.next = previous # reverse the link
            previous = current      # move `previous` one step ahead
            current = following         # move `current` one step ahead
            if following:               # if this was not the last element
                following = following.next    # move `following` one step ahead
            self.head = previous

        print(self.head.data)
            
if  __name__ == '__main__':
    lst = llist()
    lst.insert_values([4,2,2,3])
    lst.reverse()
```

```python
class Node:
    def __init__(self,data = None,next = None,prev=None):
        self.data = data
        self.next = next
        self.prev = prev
class llist:
    def __init__(self):
        self.head = None
    def insert(self,data):
        if self.head == None:
            node = Node(data,self.head,None)
            self.head = node
        else:
            node = Node(data,self.head,None)
            self.head.prev = node
            self.head = node
            
    def insert_list(self,data):
        for i in data:
            self.insert(i)
    def forword(self):
        if self.head is None:
            raise Exception("list is empty")
        itr = self.head
        llist = ""
        while itr:
            llist = llist+str(itr.data)+"<-->"
            itr = itr.next
        print(llist)
        
    def lastval(self):
        itr = self.head
        while itr:
            itr = itr.next
        return itr
            
    def back(self):
        last_node = self.lastval()
        itr = last_node
        lstr = ""
        while itr:
            lstr = lstr+itr.data+"-->"
            itr = itr.prev
        print(lstr)
        
if __name__ == "__main__":
    ls = llist()
    ls.insert(1)
    ls.insert(33)
    ls.insert(25)
    ls.insert(24)
    ls.insert(100)
    ls.back()
```
